```python
# Write a Python program to count the number of consonants in a given string.

def countConsonants(text):
    vowels = 'aeiou';
    consonants = 0;
    for i in text:
        if i not in vowels:
            consonants += 1;
            
    return consonants;
    
print(countConsonants('learning'));
```

```python
# Find the Factorial of a Number

# 1*2*3*4*5

def findFactorial(n):
    result = 1;
    for i in range(1, n+1):
        result *= i;
        
    return result;
    
print(findFactorial(5));
```

```python
# Fibonacci Sequence up to N terms

def findFibbonachiResult(n):
    result = [0,1];
    for i in range(1, n):
        result.append(result[len(result) - 1] + result[len(result) - 2])
    return result;
    
print(findFibbonachiResult(5));
```
