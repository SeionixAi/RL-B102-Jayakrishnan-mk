# 1. What is Seaborn and how is it different from Matplotlib?

Seaborn is a high-level Python visualization library built on top of Matplotlib. It provides attractive and easy-to-create statistical graphs with less code. Matplotlib offers more customization and low-level control, while Seaborn focuses on simplicity and better default styling.


| Feature           | Matplotlib          | Seaborn                   |
| ----------------- | ------------------- | ------------------------- |
| Level             | Low-level library   | High-level library        |
| Ease of Use       | More code required  | Less code required        |
| Appearance        | Basic/default style | More modern & attractive  |
| Best For          | Full customization  | Statistical visualization |
| Built On          | Independent         | Built on Matplotlib       |
| DataFrame Support | Limited/manual      | Excellent Pandas support  |
| Default Themes    | Simple              | Beautiful themes          |


# 2. What is a statistical plot?
A statistical plot is a graphical representation of statistical data used to analyze distributions, relationships, trends, and patterns in datasets. It helps in understanding data visually before performing machine learning or statistical analysis.

# 3. Difference between distplot / histplot and boxplot

In modern Seaborn,
distplot() is mostly deprecated.
Now we mainly use:
```python
sns.histplot()

```
Histplot or distplot is used to visualize the distribution and frequency of data values, while boxplot provides a statistical summary including median, quartiles, and outliers. Histplots help understand distribution shape, whereas boxplots are mainly useful for detecting spread and abnormal values.

| Feature      | Histplot / Distplot             | Boxplot                  |
| ------------ | ------------------------------- | ------------------------ |
| Purpose      | Show distribution frequency     | Show statistical summary |
| Displays     | Shape of distribution           | Quartiles & outliers     |
| Best For     | Understanding spread/pattern    | Detecting outliers       |
| Visual Style | Bars/histogram                  | Box with whiskers        |
| Detail Level | More visual distribution detail | Compact summary          |


# 4. What is a heatmap and where is it used?
A heatmap is a data visualization technique where values are represented using colors. It is commonly used to visualize correlations, missing values, patterns, and confusion matrices in machine learning and data analysis. Heatmaps help quickly identify relationships and trends in large datasets.

# 5. Why is Seaborn useful for data exploration in AI projects?
Seaborn is useful because it provides simple and attractive statistical visualizations that help understand datasets before training AI models. It helps detect patterns, outliers, missing values, feature relationships, and class imbalance during exploratory data analysis, improving data quality and model performance.
```c

Suppose you build a customer churn prediction system.

Using Seaborn you can:

    visualize customer age distribution
    detect missing salary values
    analyze relationship between subscription and churn
    identify imbalance between churn/non-churn users
    detect unusual customer behavior

Only after this,
you train the ML model.
```
