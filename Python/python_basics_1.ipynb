# even or odd--------

```python
num = int(input("Enter a number: "));
if num % 2 == 0:
    print("even");
else :
    print("odd");
```

# list ops--------

```python
numbers = [10,20,30,40,50];

print("Min: ", min(numbers))
print("Max: ", max(numbers))
print("Sum: ", sum(numbers))
```

# function example--------

```python
def square(a):
    return a*a;
    
print(square(3));
```

# loop--------

```python
for i in range(0,5):
    print(i)
```

```python
i=0;

while(i<5):
    print(i)
    i+=1;
```

# list comprehension

```python
even = [i for i in range(8) if i%2 == 0]
print(even);
```
