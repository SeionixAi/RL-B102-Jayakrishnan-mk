# 1. Dynamic Typing

```python
# Dynamic typing in Python means the data type of a variable is determined automatically at runtime. 
# Programmers do not need to explicitly declare variable types, and the same variable can hold different data types during execution. 
# This makes Python flexible and easy to use.

# Write a program that:
# * Assigns an integer value to a variable
# * Reassigns a string value to the same variable
# * Prints the value and its type at each step

value = 2;
print(value, type(value));
value = 'string';
print(value, type(value));

```

# 5. List Comprehension
List comprehension is a concise way to create lists in Python using a single line of code. 
It combines loops and optional conditions into a compact syntax, making code shorter, cleaner, and often faster than traditional loops.


```python
# Write a program that:
# * Creates a list of squares from 1 to 10

squares = [x * x for x in range(1, 10)]
print(squares)
```

# 6. break, continue, and pass
break, continue, and pass are control statements in Python. 
break terminates the loop completely, continue skips the current iteration and moves to the next one, and pass is a placeholder statement that performs no action.

```python
# Write a program that:
# * Demonstrates break, continue, and pass

for i in range(8):
    if i == 2:
        continue
    elif i % 2 == 0:
        pass
    elif i == 8:
        break
    print(i)
```

# 7. Recursion
Recursion is a programming technique where a function calls itself repeatedly to solve smaller instances of a problem. A recursive function must have a base case to stop execution; otherwise, it can lead to infinite recursion. A common example is calculating factorials.

```python
# Write a recursive function that:

def factorial(n):
    if n == 0: return 1
    return n * factorial(n - 1)

print(factorial(5))
```

# 8. Lambda Function
A lambda function is a small anonymous function in Python written using the lambda keyword. It is used for short, simple operations and can take multiple arguments but contains only a single expression. Lambda functions are commonly used with functions like map(), filter(), and sorted().

```python
# Write a lambda function that:
# * Returns the square of a number

square = lambda x: x * x
print(square(4))
```

# 9. is vs ==
== is used to compare the values of two objects, while is is used to compare their identities, meaning whether both variables refer to the same object in memory. == checks equality of content, whereas is checks memory reference equality.

```python
# Write a program that:
# * Compares two lists using == and is

a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)
print(a is b)
```

# 10. Docstrings
Docstrings are special strings written inside functions, classes, or modules using triple quotes to describe their purpose and behavior. They improve code readability, maintainability, and documentation, and can be accessed using the __doc__ attribute.

```python
# Write a function that:
# * Uses a docstring

def greet():
    """This function prints a greeting message"""
    print("Hello Python")

greet()
print(greet.__doc__)
```
