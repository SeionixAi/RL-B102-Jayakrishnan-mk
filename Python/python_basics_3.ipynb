# 2. What is the difference between read(), readline(), and readlines()?

read() reads the entire file content as a single string, readline() reads one line at a time, and readlines() reads all lines of a file and returns them as a list. These methods are used based on file size and processing requirements.

| Method        | Reads              |
| ------------- | ------------------ |
| `read()`      | Entire file        |
| `readline()`  | One line at a time |
| `readlines()` | All lines as list  |

Example
```python
with open("sample.txt", "r") as file:
    content = file.readlines()

    print(content)
```
Output:
```python
['Hello\n', 'Python\n', 'AI']
```

### When to use read()?
Use when:
file size is small , 
need complete content at once

Problem with Large Files: 
For huge files - memory usage increases.
So read() is not ideal for massive datasets/logs.

### When to use readline()?
Useful when:
processing file line-by-line , 
large files , 
streaming data , 
More memory efficient.

### Why useful - readlines()?
Easy for:
looping , 
line-wise processing

# 3. What is exception hierarchy in Python?

Exception hierarchy in Python is the tree-like structure in which all exceptions are organized using inheritance. The root class is BaseException, and most runtime errors inherit from the Exception class. Specific exceptions like ZeroDivisionError and IndexError are derived from broader parent exception classes, allowing flexible and structured error handling.

Exception Inheritance:

```python
ZeroDivisionError
        ↓
ArithmeticError
        ↓
Exception
        ↓
BaseException
```

# 4. Difference between try-except and try-except-finally
Both are used in Python: for exception handling.
try-except is used to catch and handle exceptions in Python, whereas try-except-finally includes an additional finally block that always executes whether an exception occurs or not. The finally block is commonly used for cleanup operations like closing files or releasing resources.

| Structure            | Purpose                                     |
| -------------------- | ------------------------------------------- |
| `try-except`         | Handle errors                               |
| `try-except-finally` | Handle errors + always execute cleanup code |

Main Difference
| Feature                      | `try-except` | `try-except-finally` |
| ---------------------------- | ------------ | -------------------- |
| Handles errors               | ✅ Yes        | ✅ Yes                |
| Executes cleanup code always | ❌ No         | ✅ Yes                |
| Used for resource cleanup    | ❌ Rarely     | ✅ Commonly           |


# 5. What is the purpose of the with statement?
The with statement in Python is used for automatic resource management. It ensures resources like files are properly opened and automatically closed after use, even if an exception occurs. It makes code cleaner, safer, and prevents resource leaks.

with vs Normal Approach:

# 7. What are generators and how are they different from functions?
Generators are special functions in Python that use the yield keyword to produce values one at a time instead of returning all values at once. Unlike normal functions, generators pause execution after each yield and resume later, making them memory efficient and useful for handling large datasets or streams of data.

Main Difference Between Function and Generator
| Normal Function             | Generator                |
| --------------------------- | ------------------------ |
| Uses `return`               | Uses `yield`             |
| Returns all data at once    | Produces data one-by-one |
| Stores everything in memory | Memory efficient         |
| Execution ends after return | Execution pauses/resumes |


# 8. What is *args and **kwargs?
*args and **kwargs are used in Python functions to accept variable numbers of arguments. *args collects multiple positional arguments into a tuple, while **kwargs collects multiple keyword arguments into a dictionary. They make functions more flexible and reusable.

Example
```python
def demo(*args, **kwargs):

    print(args)
    print(kwargs)

demo(1, 2, 3, name="JK", city="Kochi")
```
Output:
```python
(1, 2, 3)
{'name': 'JK', 'city': 'Kochi'}
```
