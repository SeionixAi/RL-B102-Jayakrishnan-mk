# 1. Write a Python program to count the number of vowels in a given string.

```python
def countVowels(text) :
    vowels = 'aeiou';   
    count = 0;
    for char in text.lower():
        if char in vowels:
            count += 1;
    return count;
            
print(countVowels('Artificiality'));
```

# 2. Write a Python program to check whether a given string is a palindrome.

```python
def isPalindrome(text) :
    text = text.lower();
    reversed_text = text[::-1]
    return text == reversed_text
    
print(isPalindrome('madam'));
```

# 3. Write a Python program to check whether two strings are anagrams.

```python
def isAnagram(text1, text2):
    text1 = text1.lower()
    text2 = text2.lower()
    
    text1 = sorted(text1);
    text2 = sorted(text2);
    return text1 == text2;
    
print(isAnagram('lava', 'vala'));
```
