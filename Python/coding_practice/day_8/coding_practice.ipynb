# 1. Find the Factorial of a Number

```python
def findFactorial(num):
    res = 1; 
    for i in range(1, num+1):
       res *= i;
    return res;
    
print(findFactorial(5));
```

# 2. Find the Largest and Smallest Number in a List

```python
def findBigAndSmall(listInput):
    if not listInput:
        return None;
    max = listInput[0];
    min = listInput[0];
    for i in listInput:
        if i > max:
            max = i;
        elif i < min:
            min = i;
            
    return {'Largest': max, 'Smallest': min};
print(findBigAndSmall([20,8,11,99,2]));
```

# 3. Fibonacci Sequence up to N terms

```python
def checkFibonacci(n):
    if n <= 0: 
        return [];
    if n == 1:
        return [0];
    res = [0,1];
    for i in range(n-2):
        res.append(res[-1] + res[-2])
    return res;

print(checkFibonacci(5));
```

# 4. Create a Square Frame Pattern

```python
def squareFrame(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()
    
squareFrame(5);
```

# 5. Sum of Digits in a Number

```python
def checkSum(input):
    sum = 0;
    for i in str(input):
        sum += int(i);
    return sum;
    
print(checkSum(1234));
```
