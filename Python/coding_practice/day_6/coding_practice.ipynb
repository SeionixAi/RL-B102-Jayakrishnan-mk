# 1. Write a Python program to count the number of consonants in a given string.

```python
def count_consonants(str):
    vowel = 'aeiou';
    count = 0;

    for char in str.lower():
        if char.isalpha() and char not in vowel:
            count += 1;

    return count;
print(count_consonants('Machine'));
```

# 2. Write a Python program to find the first non-repeating character in a string.

```python
def find_first_unique_char(text):
    char_count = {};
    
    for char in text.lower():
        if char in char_count:
            char_count[char] += 1;
        else:
            char_count[char] = 1;
        
    for key,val in char_count.items():
        if val == 1:
            return key;
        
    return None;
        
print(find_first_unique_char('helloh'));
```

# 3.Write a Python program to remove all duplicate characters from a string.

```python
def remove_duplicates(text):
    result = "";
    
    for char in text.lower():
        if char not in result:
            result += char;
    
    return result;
print(remove_duplicates('miami'));
```

# 4. Write a Python program to count the frequency of each character in a string.

```python
def count_frequency(text):
    frequency = {};
    
    for char in text:
        if char in frequency:
            frequency[char] += 1;
        else:
            frequency[char] = 1;
            
    return frequency;
print(count_frequency('heeyoo'));
```
