```julia
# 1. Dynamic Typing
# Dynamic typing in Python means the data type of a variable is determined automatically at runtime. 
# Programmers do not need to explicitly declare variable types, and the same variable can hold different data types during execution. 
# This makes Python flexible and easy to use.

# Write a program that:
# * Assigns an integer value to a variable
# * Reassigns a string value to the same variable
# * Prints the value and its type at each step

value = 2;
print(value, type(value));
value = 'string';
print(value, type(value));

```

# 5. List Comprehension
List comprehension is a concise way to create lists in Python using a single line of code. 
It combines loops and optional conditions into a compact syntax, making code shorter, cleaner, and often faster than traditional loops.


```julia
# Write a program that:
# * Creates a list of squares from 1 to 10

squares = [x * x for x in range(1, 10)]
print(squares)
```
