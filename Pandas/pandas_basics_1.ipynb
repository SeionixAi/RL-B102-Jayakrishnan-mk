# Pandas Operations

```python
import pandas as pd;

# Create DataFrame

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [22, 24, 23],
    "Marks": [85, 90, 88]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Basic Operations

print(df["Age"].mean())
print(df["Marks"].max())
```

```python
# Filtering
print(df[df["Marks"] > 85])
```
