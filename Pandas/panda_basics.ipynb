# 1. What is data cleaning in Pandas?
Data cleaning in Pandas is the process of identifying and correcting missing, incorrect, duplicate, or inconsistent data before analysis or Machine Learning. Pandas provides functions like dropna(), fillna(), drop_duplicates(), and astype() to efficiently clean and preprocess datasets.

# 2. What is the difference between loc and iloc?
loc is a label-based indexing method in Pandas used to access rows and columns using row labels and column names, while iloc is an integer position-based indexing method used to access data using numerical indices. loc includes the ending index in slicing, whereas iloc follows normal Python slicing rules and excludes the ending index.

### Quick Difference Table
| Feature               | `loc`        | `iloc`              |
| --------------------- | ------------ | ------------------- |
| Selection type        | Label-based  | Position-based      |
| Uses row/column names | ✅ Yes        | ❌ No                |
| Uses index numbers    | ❌ Not mainly | ✅ Yes               |
| Inclusive slicing     | ✅ Yes        | ❌ No (Python-style) |


# 3. What are missing values and how does Pandas handle them?
Missing values are empty or unavailable data entries in a dataset, usually represented as NaN or None in Pandas. 
Pandas provides functions like isnull(), dropna(), and fillna() to detect, remove, or replace missing values during data cleaning and preprocessing.

| Method              | Purpose                 |
| ------------------- | ----------------------- |
| `dropna()`          | Remove missing data     |
| `fillna()`          | Replace missing data    |
| Mean/Median filling | Statistical replacement |
| Forward fill        | Use previous value      |
| Backward fill       | Use next value          |


# 4. What is groupby() and why is it used?
groupby() in Pandas is used to group rows with similar values together and perform aggregate operations like sum, mean, count, max, or min on each group. It is widely used for data analysis, summarization, and preprocessing in Data Science and AI workflows.

### Common Operations with groupby()
| Function  | Purpose    |
| --------- | ---------- |
| `mean()`  | Average    |
| `sum()`   | Total      |
| `count()` | Count rows |
| `max()`   | Maximum    |
| `min()`   | Minimum    |
