```python
# 1. Importing Pandas

import pandas as pd

# 2. Load Data

df=pd.read_csv('data.csv') # Load CSV
df = pd.read_excel('data/data.xlsx') # Load Excel
```

```python
import numpy as np
np.random.seed(1)
rows = 200

df = pd.DataFrame({
    "PassengerId": np.arange(1, rows + 1),
    "Name": np.random.choice(["Mr. John", "Mrs. Anna", "Miss Emma", "Dr. Smith"], rows),
    "Age": np.random.choice(np.append(np.random.randint(1, 80, 180), [np.nan]*20), rows),
    "Sex": np.random.choice(["male", "female"], rows),
    "Pclass": np.random.choice([1, 2, 3], rows),
    "Fare": np.round(np.random.uniform(10, 500, rows), 2)
})

# 3. View Data

print(df.head());
```

```python
print(df.tail());
```

```python
print(df.sample());
```

```python
# 5. Handling Missing Data

print(df.isnull().sum())
```

```python
print(df.dropna(inplace=True))
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
# 6. Removing Duplicates

df.duplicated().sum()
```

```python
print(df.drop_duplicates(inplace=True));
```

```python
# 7. Selecting Columns
print(df['Name']);
```

```python
print(df[['Name', 'Age']]);
```

```python
# 8. Filtering Rows
df[df['Age'] > 30]
```

```python
df[df['Sex'] == 'female']
```

```python
df[(df['Age'] > 30) & (df['Sex'] == 'female')]
```

```python
# 9. Sorting Data
df.sort_values(by='Age', ascending=True)
```

```python
df.sort_values(by=['Pclass', 'Fare'], ascending=[True, False])
```

```python
# 10. Creating New Columns

df['UniqueId'] = df['PassengerId']  + df['Age'] + 1
print(df.head())
```

```python
# 11. Dropping Columns or Rows

df.drop('UniqueId', axis=1, inplace=True);
print(df.head())
```

```python
df.drop([0,2], axis=0, inplace=True);
print(df.head())
```

```python
# 12. Renaming Columns

df.rename(columns={'Fare': 'FareRate'}, inplace=True);
print(df.head())
```

```python
# 13. Value Counts & Unique

print(df['Sex'].value_counts());
```

```python
print(df['Name'].unique())
```

```python
print(df['Name'].nunique())
```

```python
# 14. Grouping and Aggregation

```

```python
print(df.groupby('Pclass')['FareRate'].mean())
```

```python
print(df.groupby('Sex')['FareRate'].sum())
```

```python
print(df.groupby(['Pclass', 'Sex'])['Age'].median())
```

```python
# 15. Aggregating Multiple Columns

# group by pclass and find the mean and min of age and mean and max of fare.
print(df.groupby('Pclass').agg({
    'Age': ['mean', 'min'],
    'FareRate': ['mean', 'max']
}))
```

```python
# 16. Apply Custom Functions

df['AgeSquared'] = df['Age'].apply(lambda x: x**2);
print(df)
```

```python
# 17. String Operations
print(df['Name'].str.lower())
```

```python
print(df['Name'].str.contains('Mrs'))
```

```python
# 18. Mapping & Replacing Values

df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
print(df.head())
```

```python
df['Name'].replace({'Mrs. Anna': "Mrs. Ann", 'Mr. John': 'Johnny'}, inplace=True);
print(df.head())
```

```python
# 19. Combining Columns (concatenation)

df['fullname'] = df['Name'] + ' (' + df['Sex'].astype(str) + ')'
print(df.head());
```
