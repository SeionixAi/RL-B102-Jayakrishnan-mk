# DATASET 1: Titanic-Like Passenger Dataset
### Purpose: Learning data inspection, filtering, missing values, grouping

```python
# Create Dataset using NumPy -

import numpy as np
import pandas as pd

np.random.seed(1)
rows = 200

df = pd.DataFrame({
    "PassengerId": np.arange(1, rows + 1),
    "Name": np.random.choice(["Mr. John", "Mrs. Anna", "Miss Emma", "Dr. Smith"], rows),
    "Age": np.random.choice(np.append(np.random.randint(1, 80, 180), [np.nan]*20), rows),
    "Sex": np.random.choice(["male", "female"], rows),
    "Pclass": np.random.choice([1, 2, 3], rows),
    "Fare": np.round(np.random.uniform(10, 500, rows), 2)
})

# View Data Methods-

print(df.head()); # head - view first 5 rows
```

```python
print(df.tail()); # tail - view last 5 rows
```

```python
print(df.sample(3)); # sample - view random 3 rows
```

```python
# Inspect Structure Methods-

print(df.shape)   # Number of rows and columns
```

```python
print(df.columns)  # Column names
```

```python
print(df.info())  # Data types and missing values
```

```python
print(df.describe()) # Statistical summary
```

```python
# Handle Missing Data Methods -

print(df.isnull().sum());   # Count missing values
```

```python
print(df["Age"].fillna(df["Age"].mean())); # filling the average age in missing rows
```

```python
# Filtering Rows Methods -

print(df[df["Age"] > 30]) # filter by age > 30
```

```python
print(df[df["Sex"] == 'female']); # filter by sex - female
```

```python
# Grouping & Aggregation -

print(df.groupby("Pclass")["Fare"].mean()) # group by Pclass and averageing the fare.
```
