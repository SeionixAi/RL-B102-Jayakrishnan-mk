# 1. What is Matplotlib and why is it used?
Matplotlib is a Python library used for data visualization and plotting graphs such as line charts, bar charts, pie charts, histograms, and scatter plots. It is widely used in Data Science and AI to visualize datasets, analyze trends, and understand model performance.

Common Types of Graphs
| Graph Type   | Purpose                        |
| ------------ | ------------------------------ |
| Line Plot    | Trends over time               |
| Bar Chart    | Compare categories             |
| Pie Chart    | Show proportions               |
| Scatter Plot | Relationship between variables |
| Histogram    | Distribution of data           |


# 2. Difference between line plot, bar plot, and scatter plot

A line plot is used to show trends or continuous changes by connecting data points with lines. A bar plot is used to compare categories using rectangular bars. A scatter plot displays individual points to analyze relationships or correlations between two variables.

| Plot Type    | Best Used For                  | Visual Style      |
| ------------ | ------------------------------ | ----------------- |
| Line Plot    | Trends over time               | Connected lines   |
| Bar Plot     | Comparing categories           | Rectangular bars  |
| Scatter Plot | Relationship between variables | Individual points |


# 3. What are figure and axes in Matplotlib?
In Matplotlib, a Figure is the top-level container that represents the entire plotting window or canvas, while Axes is the actual area inside the figure where data is plotted. A figure can contain one or more axes, and most plotting operations are performed on axes objects.

```python
Figure → whole canvas
Axes   → actual plotting area
```

# 4. How do labels, titles, and legends improve data visualization?

Titles provide overall context and describe what the graph represents.

Labels explain what the x-axis and y-axis represent, making the graph understandable.

Legends identify different datasets, colors, or lines in a graph.

```python
Internally Matplotlib stores:

axis metadata
label strings
legend handles

Then while rendering:

title placed top
labels near axes
legend box generated automatically

```
So visualization is not just “drawing”.
It’s structured communication.
