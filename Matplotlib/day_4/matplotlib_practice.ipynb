```python
import matplotlib.pyplot as plt;
import numpy as np;
```

```python
# sample data 
x = np.linspace(0, 10, 100);
y = np.sin(x);

# Line plot
plt.plot(x, y)
plt.title('Line plot');
plt.show()
```

```python
# 2. Scatter plot
plt.scatter(x, y)
plt.title("Scatter Plot")
```

```python
# bar chart
categories = ['a', 'b', 'c'];
values = [10, 20, 15];
plt.bar(categories, values);
plt.title("Bar Chart")
```

```python
# horizontal bar chart
plt.barh(categories, values);
plt.title("Horizontal Bar Chart")
```

```python
# Histogram
data = np.random.randn(1000);
plt.hist(data, bins=30)
plt.title("Histogram")
```

```python
plt.boxplot(data);
plt.title("Box Plot")
```

```python
# pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title("Pie Chart")
```

```python
# 8. Multiple lines on one plot
plt.plot(x, np.sin(x), label='sin');
plt.plot(x, np.cos(x), label='cos');
plt.legend()
plt.title("Multiple Lines")
```

```python
# 9. Subplots
fig, axs = plt.subplots(2,1);
axs[0].plot(x, np.sin(x));
axs[1].plot(x, np.cos(x));
plt.suptitle("Subplots Example")
```

```python
# 10. Changing color and style
plt.plot(x, y, color='red', linestyle='--', marker='o');
plt.title('Styled line');
```

```python
# 11. Adding labels and title
plt.plot(x, y)
plt.xlabel("X Axis");
plt.ylabel("Y Axis");
plt.title("Labeled Plot")
```

```python
# 12. Fill between
plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.title("Fill Between")
```

```python
# 13. Annotate points
plt.plot(x, y)
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(2, 1.2), arrowprops=dict(facecolor='black', shrink=0.05));
plt.title("Annotation Example")
```

```python
# 14. Save figure
plt.plot(x, y)
plt.savefig('line_plot.png');
```

```python
# 15. Gridlines
plt.plot(x, y)
plt.grid(True);
plt.title("Gridlines Example")
```
