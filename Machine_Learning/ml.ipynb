# 1 Machine Learning: A Brief Overview
```python
Machine Learning is a subset of Artificial Intelligence that enables 
computers to learn from data and improve their performance on tasks 
without being explicitly programmed.

Key concepts:
- Algorithms learn patterns from training data
- Three main types: Supervised, Unsupervised, and Reinforcement Learning
- Used for predictions, classifications, clustering, and pattern recognition
- Powers applications like recommendation systems, image recognition, and NLP
```


# 2 What are the main types of Machine Learning?
```python
The three main types of Machine Learning are :
    Supervised Learning, 
    Unsupervised Learning, 
    Reinforcement Learning. 

Supervised Learning uses labeled data to make predictions, 
Unsupervised Learning finds hidden patterns in unlabeled data, and 
Reinforcement Learning learns optimal actions through rewards and penalties by interacting with an environment.

Comparison Table:
```
| Type                   | Training Data       | Goal                 | Example                |
| ---------------------- | ------------------- | -------------------- | ---------------------- |
| Supervised Learning    | Labeled data        | Predict outputs      | House price prediction |
| Unsupervised Learning  | Unlabeled data      | Find hidden patterns | Customer segmentation  |
| Reinforcement Learning | Rewards & penalties | Learn best actions   | Robot learning         |

# 3 What is the difference between Classification and Regression?
```python
Classification predicts discrete categories or labels, such as spam/not spam or pass/fail, 
while Regression predicts continuous numerical values, such as house prices or salaries. 

Both are supervised learning techniques, 
    but Classification is used for categorization and 
    Regression is used for value prediction.

Comparison Table
```
| Feature        | Classification   | Regression        |
| -------------- | ---------------- | ----------------- |
| Output Type    | Category/Label   | Continuous Number |
| Goal           | Assign a class   | Predict a value   |
| Example        | Spam or Not Spam | House Price       |
| Output Example | Yes / No         | ₹50,00,000        |
| Used For       | Categorization   | Value prediction  |


# 4 What is a feature in Machine Learning?

```python
A feature is an input variable or attribute used by a Machine Learning model to learn patterns and make predictions. 
Features represent the information provided to the model, while the value being predicted is called the target or output variable. Good feature selection is essential for building accurate and effective ML models.

Types of Features:

    Numerical Features - Contain numbers.
        Examples:
            Age
            Salary
            Height
            Weight

    Categorical Features - Contain categories or labels.
        Examples:
            Gender
            City
            Blood Group
            Product Category

These often need to be converted into numbers before training a model.
```

# 5 What is a label or target variable?

```python
A label or target variable is the output that a Machine Learning model is trained to predict. 
It represents the correct answer in the training data, while the input variables used to make the prediction are called features. 
In supervised learning, the model learns the relationship between features and the target variable to make predictions on new data.
```

# 1. What are Overfitting and Underfitting?

```python
Underfitting occurs when a model is too simple to capture the underlying patterns in the data, resulting in poor performance on both training and testing datasets.
Overfitting occurs when a model learns the training data too closely, including noise and outliers, leading to excellent training performance but poor performance on unseen data.
```
| Feature           | Underfitting | Overfitting      |
| ----------------- | ------------ | ---------------- |
| Model Complexity  | Too simple   | Too complex      |
| Training Accuracy | Low          | Very high        |
| Testing Accuracy  | Low          | Low              |
| Learns Patterns   | No           | Learns noise too |
| Generalization    | Poor         | Poor             |


# 2. What is the difference between Supervised and Unsupervised Learning?
```python

Supervised Learning trains a model using labeled data, where both inputs and correct outputs are available, making it suitable for tasks like classification and regression. 
Unsupervised Learning uses unlabeled data, where the model identifies hidden patterns or groups in the data without predefined outputs, making it useful for clustering and pattern discovery.
```
| Feature          | Supervised Learning        | Unsupervised Learning   |
| ---------------- | -------------------------- | ----------------------- |
| Training Data    | Labeled                    | Unlabeled               |
| Output Available | Yes                        | No                      |
| Goal             | Predict output             | Find hidden patterns    |
| Learns From      | Features + Labels          | Features only           |
| Common Tasks     | Classification, Regression | Clustering, Association |
| Example          | Spam detection             | Customer segmentation   |


# 3. What is a training dataset and a testing dataset? Why is data splitting important?

```python
A training dataset is used to teach a Machine Learning model by allowing it to learn patterns from the data. 
A testing dataset is a separate set of unseen data used to evaluate the model's performance. 
Data splitting is important because it measures how well the model generalizes to new data, helps detect overfitting, and provides an unbiased evaluation of the model's accuracy and reliability.

                Total Dataset
                     │
        ┌────────────┴────────────┐
        │                         │
Training Dataset          Testing Dataset
   (Learn)                  (Evaluate)
        │                         │
        └────────────┬────────────┘
                     │
          Measure Model Performance

```

# 4. What is feature scaling and why is it needed in some algorithms?

```python
Feature scaling is the process of transforming features to a similar scale so that no feature dominates due to its larger numerical values. 
It is especially important for distance-based and gradient-based algorithms such as KNN, K-Means, SVM, Logistic Regression, and Neural Networks, as it improves model accuracy, speeds up training, and ensures fair contribution from all features.
```
